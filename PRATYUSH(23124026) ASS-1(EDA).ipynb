I am working with DATASET Financial Statements of Major Companies(2009-2023) and perform Exploratory Data Analysis(EDA) in two steps:

1. Univariate Analysis (single variable insights)
    - Revenue distribution, profit margins, expense trends.
    - Summary statistics (mean, median, mode)
    - Histograms, boxplots, and bar charts 

2. Multivariate Analysis (relationships between variables)
    - Correlation Between Revenue, Profit, and Expenses
    -  Scatter Plot of Revenue vs Profit
    -  Boxplot of Profit by Sector



### importing the dataset

```python
 import pandas as pd

# Load the dataset

file_path = "Financial Statements of Major Companies(2009-2023).csv"
df = pd.read_csv(file_path)

# Display first five rows
df.head()
```

```python
# Check dataset structure and missing values
df.info()

# Summary statistics of numerical columns
df.describe()
```

### Cleaning the data

```python
# Identify missing values in each column
missing_values = df.isnull().sum()
print(missing_values[missing_values > 0])  # Show only columns with missing values

# Separate numerical and categorical columns
numerical_cols = df.select_dtypes(include=["number"]).columns
categorical_cols = df.select_dtypes(include=["object"]).columns

# Fill missing values
df[numerical_cols] = df[numerical_cols].fillna(df[numerical_cols].mean())  # Fill numerical columns with mean
df[categorical_cols] = df[categorical_cols].fillna(df[categorical_cols].mode().iloc[0])  # Fill categorical columns with most common value

# Check again if any missing values remain
print(df.isnull().sum().sum())  # Should print 0 if all missing values are filled
```

### Univariate Analysis (Single Variable Exploration)
a) Distribution of Revenue

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram of revenue
plt.figure(figsize=(8,5))
sns.histplot(df["Revenue"], bins=30, kde=True)
plt.title("Distribution of Revenue")
plt.xlabel("Revenue")
plt.ylabel("Frequency")
plt.show()
```

```python
print(df.columns)
```

b) Count of Companies per Sector

```python
plt.figure(figsize=(10, 5))
sns.countplot(y=df["Category"], order=df["Category"].value_counts().index, palette="RdYlGn_r")
plt.title("Number of Companies per Category")
plt.xlabel("Count")
plt.ylabel("Category")
plt.show()
```

### Calculating the total expenses

```python
# Assuming Net Expenses = Revenue - Gross Profit
df["Net Expenses"] = df["Revenue"] - df["Gross Profit"]

# Display the first few rows to check the result
print(df[["Revenue", "Gross Profit", "Net Expenses"]].head())
```

### Multivariate Analysis (Exploring Relationships)
a) Correlation Between Revenue, Profit, and Expenses

```python
# Compute correlation matrix
correlation_matrix = df[["Revenue", "Gross Profit", "Net Expenses"]].corr()

# Heatmap of correlations
plt.figure(figsize=(6, 5))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

b) Scatter Plot of Revenue vs Profit

```python
# Scatter plot between Revenue and Profit
plt.figure(figsize=(8,5))
sns.scatterplot(x=df["Revenue"], y=df["Gross Profit"], color="green")
plt.title("Revenue vs Gross Profit")
plt.xlabel("Revenue")
plt.ylabel("Gross Profit")
plt.show()
```

c) Boxplot of Profit by Sector

```python
# Boxplot to see Profit distribution by sector
plt.figure(figsize=(6,6))
sns.boxplot(x=df["Category"], y=df["Gross Profit"])
plt.xticks(rotation=90)
plt.title("Profit Distribution Across Sectors")
plt.show()
```
